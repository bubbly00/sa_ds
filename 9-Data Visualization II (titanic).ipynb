```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

```python
dataset = sns.load_dataset('titanic')
tips = load_dataset("tips")

dataset.head()
```

```python
tips.head()
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived");
```

```python
print("If we want to see the box plots of forage of passengers of both genders, along with the information about whether or not they survived, we can pass the survived as value to the hue parameter. We can also see the distribution of the passengers who survived. For instance, we can see that among the male passengers, on average more younger people survived as compared to the older ones. Similarly, we can see that the variation among the age of female passengers who did not survive is much greater than the age of the surviving female passengers.")
```
